Collaborative Filtering Based Recommendation System

```python
TUsers = book_ratings.groupby('User-ID').count()['Book-Rating'] > 100
trusted_users = TUsers[TUsers].index
```

```python
filtered_rating = book_ratings[book_ratings['User-ID'].isin(trusted_users)]
```

```python
fbooks = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = fbooks[fbooks].index
```

```python
recommender_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
t = recommender_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
t.fillna(0,inplace=True)
```

```python
t
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(t)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(t.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    book = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == t.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        book.append(item)
    
    return book
```

```python
recommend('24 Hours')
```
